```python
import pandas as pd

# Load CSVs
users = pd.read_csv("expanded_feature_users_dataset (1).csv", parse_dates=["signup_date"])
feature = pd.read_csv("expanded_feature_usage_dataset.csv", parse_dates=["feature_usage_date"])

# Create a flag column: Did this user use the feature?
users["used_feature"] = users["user_id"].isin(feature["user_id"])

# Save final dataset to a new CSV file
users.to_csv("final_users_with_used_feature.csv", index=False)

print("✅ Dataset with 'used_feature' created successfully!")

```

```python
# Add revenue growth column
users["revenue_growth"] = users["revenue_after"] - users["revenue_before"]
print("✅ Revenue growth column added.")
```

```python
# Summary by feature usage
summary = users.groupby("used_feature")[["revenue_before", "revenue_after", "revenue_growth"]].mean().round(2)
print("📊 Revenue Summary by Feature Usage:")
print(summary)
```

```python
# Save again (final version)
users.to_csv("final_users_with_used_feature.csv", index=False)
print("💾 Final file saved for Tableau dashboard!")

```
